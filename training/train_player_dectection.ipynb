```python
ROBOFLOW_API_KEY = userdata.get('ROBOFLOW_API_KEY')
rf = Roboflow(api_key=ROBOFLOW_API_KEY)

project = rf.workspace("work-wzbto").project("football-players-detection-mxxoa")
version = project.version(1)
dataset = version.download("yolov8")
```

```python
model = YOLO("yolov8s.pt")

results = model.train(
    data="/content/datasets/Football-Players-Detection-1/data.yaml",
    epochs=50,
    imgsz=1028,
    batch=6
)
```
